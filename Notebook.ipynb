```r
# Instalar pacotes necessários
if (!require("dplyr")) install.packages("dplyr")
if (!require("readr")) install.packages("readr")
if (!require("ggplot2")) install.packages("ggplot2")
if (!require("reshape2")) install.packages("reshape2")

# Carregar pacotes necessários
library(dplyr)
library(readr)
library(ggplot2)
library(reshape2)

# Gerar dados de exemplo (substitua pelos seus dados reais)
set.seed(123)
n <- 100
Participant <- factor(rep(1:20, each = 5))
Condition <- factor(rep(c("A", "B"), times = 50))
EDA_mean <- rnorm(n, mean = 10, sd = 2)
ECG_HR <- rnorm(n, mean = 70, sd = 5)
EEG_alpha <- rnorm(n, mean = 8, sd = 1)
EEG_beta <- rnorm(n, mean = 15, sd = 3)
EyeTracking_Fixations <- rpois(n, lambda = 5)
STAI_score <- rnorm(n, mean = 40, sd = 10)
SPSES_score <- rnorm(n, mean = 30, sd = 8)

data <- data.frame(Participant, Condition, EDA_mean, ECG_HR, EEG_alpha,
                   EEG_beta, EyeTracking_Fixations, STAI_score, SPSES_score)

# Boxplot de EDA_mean por Condição
ggplot(data, aes(x = Condition, y = EDA_mean, fill = Condition)) +
  geom_boxplot() +
  labs(title = "Boxplot de EDA Média por Condição", x = "Condição", y = "EDA Média") +
  theme_minimal()

# Boxplot de ECG_HR por Condição
ggplot(data, aes(x = Condition, y = ECG_HR, fill = Condition)) +
  geom_boxplot() +
  labs(title = "Boxplot de Frequência Cardíaca por Condição", x = "Condição", y = "Frequência Cardíaca (bpm)") +
  theme_minimal()

# Boxplot de EEG_alpha por Condição
ggplot(data, aes(x = Condition, y = EEG_alpha, fill = Condition)) +
  geom_boxplot() +
  labs(title = "Boxplot de EEG Alpha por Condição", x = "Condição", y = "EEG Alpha") +
  theme_minimal()

# Boxplot de EEG_beta por Condição
ggplot(data, aes(x = Condition, y = EEG_beta, fill = Condition)) +
  geom_boxplot() +
  labs(title = "Boxplot de EEG Beta por Condição", x = "Condição", y = "EEG Beta") +
  theme_minimal()

# Boxplot de STAI_score por Condição
ggplot(data, aes(x = Condition, y = STAI_score, fill = Condition)) +
  geom_boxplot() +
  labs(title = "Boxplot de STAI Score por Condição", x = "Condição", y = "STAI Score") +
  theme_minimal()

# Boxplot de SPSES_score por Condição
ggplot(data, aes(x = Condition, y = SPSES_score, fill = Condition)) +
  geom_boxplot() +
  labs(title = "Boxplot de SPSES Score por Condição", x = "Condição", y = "SPSES Score") +
  theme_minimal()

# Heatmap da Matriz de Correlação
correlation_matrix <- cor(data[, 3:9], method = "pearson")
cor_matrix_melted <- melt(correlation_matrix)
ggplot(cor_matrix_melted, aes(x = Var1, y = Var2, fill = value)) +
  geom_tile() +
  scale_fill_gradient2(low = "blue", mid = "white", high = "red", midpoint = 0,
                       limit = c(-1, 1), name = "Correlação") +
  labs(title = "Mapa de Calor da Matriz de Correlação") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

# Gráfico de Densidade de EDA_mean por Condição
ggplot(data, aes(x = EDA_mean, fill = Condition)) +
  geom_density(alpha = 0.5) +
  labs(title = "Densidade de EDA Média por Condição", x = "EDA Média", y = "Densidade") +
  theme_minimal()

# Gráfico de Densidade de ECG_HR por Condição
ggplot(data, aes(x = ECG_HR, fill = Condition)) +
  geom_density(alpha = 0.5) +
  labs(title = "Densidade de Frequência Cardíaca por Condição", x = "Frequência Cardíaca (bpm)", y = "Densidade") +
  theme_minimal()

# Gráfico de Densidade de EEG_alpha por Condição
ggplot(data, aes(x = EEG_alpha, fill = Condition)) +
  geom_density(alpha = 0.5) +
  labs(title = "Densidade de EEG Alpha por Condição", x = "EEG Alpha", y = "Densidade") +
  theme_minimal()

# Gráfico de Densidade de EEG_beta por Condição
ggplot(data, aes(x = EEG_beta, fill = Condition)) +
  geom_density(alpha = 0.5) +
  labs(title = "Densidade de EEG Beta por Condição", x = "EEG Beta", y = "Densidade") +
  theme_minimal()

# Gráfico de Densidade de STAI_score por Condição
ggplot(data, aes(x = STAI_score, fill = Condition)) +
  geom_density(alpha = 0.5) +
  labs(title = "Densidade de STAI Score por Condição", x = "STAI Score", y = "Densidade") +
  theme_minimal()

# Gráfico de Densidade de SPSES_score por Condição
ggplot(data, aes(x = SPSES_score, fill = Condition)) +
  geom_density(alpha = 0.5) +
  labs(title = "Densidade de SPSES Score por Condição", x = "SPSES Score", y = "Densidade") +
  theme_minimal()
```
